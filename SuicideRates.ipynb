###Grafico con matplotlib

````python

import matplotlib.pyplot as plt

plt.scatter(df['Year'], df['SuicideCount'])

plt.title('Suicide Count vs Year')
plt.xlabel('Year')
plt.ylabel('Suicide Count')

plt.show()
````

###Grafico con *seaborn*

````python

import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x="SuicideCount", y="Sex")

plt.title('Suicide Count vs Sex')
plt.xlabel('Suicide Count')
plt.ylabel('Sex')

plt.show()
````

###Graficos con plotly



````python

import plotly.express as px

fig = px.scatter(df, x="SuicideCount", y="RegionName")
fig.show()
````

````python

fig = px.pie(df, values='SuicideCount', names='RegionName',
             title='Relationship between Suicide Count and Region Name')
fig.show()
````
