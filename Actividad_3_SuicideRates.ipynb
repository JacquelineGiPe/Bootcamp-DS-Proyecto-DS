# **Visualizacion univariada**

```python

import matplotlib.pyplot as plt

plt.scatter(df['Year'], df['SuicideCount'])

plt.title('Suicide Count vs Year')
plt.xlabel('Year')
plt.ylabel('Suicide Count')

plt.show()
```


**Análisis de la Tendencia:**

La visualización presenta una tendencia compleja en las tasas de suicidio durante el período de 1990 a 2022. Se observan tres fases distintivas:

**Fase 1 (1990-1993)**: Se caracteriza por un aumento gradual en las tasas de suicidio.

**Fase 2 (1993-2003)**: Se presenta un pico significativo en las tasas de suicidio, seguido de una estabilización hasta el año 2003.

**Fase 3 (2004-2022)**: Se observa una disminución gradual en las tasas de suicidio hasta el año 2010, seguida de un nuevo incremento gradual hasta el 2022.

**Conclusion**:

La visualización presenta una tendencia compleja en las tasas de suicidio durante el período analizado. Se observan tres fases con diferentes patrones de aumento, estabilización y disminución. La interpretación precisa de la tendencia y la identificación de las causas subyacentes requieren un análisis más profundo de los datos y la consideración de diversos factores contextuales.

# **Visualizacion bivariada**


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x="SuicideCount", y="Sex")

plt.title('Suicide Count vs Sex')
plt.xlabel('Suicide Count')
plt.ylabel('Sex')

plt.show()
```

**Interpretacion**: La cantidad de suicidios masculinos es cinco veces mayor que la de suicidios femeninos. Esta disparidad significativa es un hallazgo preocupante que refleja las complejas diferencias en los factores de riesgo y las experiencias de salud mental entre hombres y mujeres.

**Factores socioculturales**: Las normas de género, las expectativas sociales y los roles de género pueden contribuir a que los hombres sean menos propensos a buscar ayuda para problemas de salud mental y más propensos a recurrir a métodos de suicidio letales.
